# Numpy


Create a null vector of size 10.

```python
import numpy as np

zeros = np.zeros(10)
zeros[4] = 1
print(zeros)
```

Create a 8x8 matrix and fill it with a checkerboard pattern.

```python
board = np.zeros((8,8))
board[1::2, ::2] = 1
board[::2, 1::2] = 1
print(board)
```
