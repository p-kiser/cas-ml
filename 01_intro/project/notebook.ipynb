# Import

```python
import pandas as pd
```

```python
imp = pd.read_csv('data/dogs-importBreeds.csv', sep=';', skiprows=1)
imp.shape
```

```python
imp.head()
```

```python
imp.isnull().sum()
```

```python
dog_cols = imp.columns.drop(['Year', 'Month'])
dog_cols
```

```python
imp[dog_cols].sum()
```

```python
dog_sums = pd.DataFrame({'rasse': dog_cols, 'sum': imp[dog_cols].sum()})
dog_sums.reset_index()
```

```python
top_dogs = dog_sums['sum'].sort_values(ascending=False).head(10)
top_dogs
```

```python
import matplotlib.pyplot as plt
```

# IMPORT

```python
import math
_ = plt.pie(
    top_dogs, 
    labels=[f"{x} ({math.floor(top_dogs[x])})" for x in top_dogs.keys()],
    colors=plt.color_sequences.get('Set3'),
    shadow=True,
    explode=(0.55,0.34,0.21,0.13,0.08,0.05,0.03,0.02,0.01,0.01),
    autopct='%1.1f%%',
)
```

# EXPORT

```python
exp = pd.read_csv('data/dogs-exportBreeds.csv', sep=';', skiprows=1)
exp.shape
exp.isnull().sum()
dog_cols2 = exp.columns.drop(['Year', 'Month'])
dog_cols2

exp[dog_cols2].sum()
dog_sums2 = pd.DataFrame({'rasse': dog_cols2, 'sum': exp[dog_cols2].sum()})
dog_sums2.reset_index()
top_dogs2 = dog_sums2['sum'].sort_values(ascending=False).head(10)


plt.pie(
    top_dogs2, 
    labels=[f"{x} ({math.floor(top_dogs2[x])})" for x in top_dogs2.keys()],
    colors=plt.color_sequences.get('Set3'),
    shadow=True,
    explode=(0.55,0.34,0.21,0.13,0.08,0.05,0.03,0.02,0.01,0.01),
    autopct='%1.1f%%'
)

plt.show()
```

```python
plt.bar(top_dogs.keys(),height=top_dogs.values)
plt.title("Top dawgs")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
dog_sums_merge = dog_sums.merge(dog_sums2, on='rasse')
dog_sums_merge.rename(columns={"sum_x": "export", "sum_y": "import"}, inplace=True)
dog_sums_merge
# pd.DataFrame({"export": dog_sums, "import": dog_sums2})
```

```python
dog_sums_both_values = dog_sums_merge[
        (dog_sums_merge["export"] > 0) & (dog_sums_merge["import"] > 0)
]
dog_sums_both_values.set_index("rasse",inplace=True)
dog_sums_both_values
```

```python
dog_sums_both_values['diff'] = dog_sums_both_values['export'] - dog_sums_both_values['import']
dog_sums_both_values['diff'].sort_values()
dog_sums_top = dog_sums_both_values.sort_values('diff').head(15)
dog_sums_top
```

```python
plt.figure('bar', figsize=(11,5))
plt.barh(dog_sums_top.index, dog_sums_top["export"], align='edge', height=0.3)
plt.barh(dog_sums_top.index,dog_sums_top["import"], height=-0.3)
plt.xticks(rotation=90, ha='right')
plt.show()
```
