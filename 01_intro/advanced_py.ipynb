From the two following lists: `x = [2,4,5,6,7,1,6,8]`, `y = [3,4,5,6,0,3,9]` create a new list with the non common elements using sets.

```python
x = [2,4,5,6,7,1,6,8]
y = [3,4,5,6,0,3,9]

# list((set(x) - set(y)) | (set(y) - set(x)))
# list(set(x).symmetric_difference(set(y)))
list(set(x) ^ set(y))
```

Make a list of the squares of the even numbers between 1 to 10. Convert your code to a list comprehension afterwards.

```python
squares = []
for i in range(1,11):
    squares.append(i**2)

squares
```

```python
[x**2 for x in range(1,11)]
```

Create a function that takes a list of numbers and returns a list with only the ones that are even.

```python
def even(input: list) -> list:
    return list(filter(lambda x: x % 2 == 0, input))

even(range(0, 20))
```

Multiply by 4 each element from a list and store it in another list. Use list comprehension.

```python
lst = [2,4,5,7,1,5,7,4,2,9,8,6]

[x * 4 for x in lst]
```

Use list comprehension to filter numbers above 60 and append them to
another list

```python
lst = [66,53,23,67,77,86,88,34,78,99,100,45,93]
[x for x in lst if x > 60]
```

Create a new dictionary with converted values to CHF (swiss francs, rate 1 EUR = 1.18 CHF). Use dictionary comprehension.

```python
d = {'milk': 1.65, 'coffee': 4.5, 'bread': 2.75, 'sugar': 0.55}

{ key:value * 1.18 for key, value in d.items()}
```
